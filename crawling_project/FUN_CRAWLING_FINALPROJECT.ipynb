#### 엑셀 기본 코드
```python
import requests, openpyxl
from bs4 import BeautifulSoup

excel_file = openpyxl.Workbook()

# active 시트 이름 변경

excel_file.save('MYSITE.xlsx')
excel_file.close()
```
    
#### 사용할 주요 openpyxl 기능 

- **Workbook 생성**: `openpyxl.Workbook()`으로 새 Workbook 생성

    ```python
    workbook = openpyxl.Workbook()
    ```

- **Worksheet 이름 변경**: `worksheet.title = 'New Title'`로 Worksheet 이름 변경

    ```python
    worksheet = workbook.active
    worksheet.title = 'Sheet1'
    ```

- **데이터 추가**: `worksheet.append([data1, data2, data3, ...])`로 새 행 추가

    ```python
    worksheet.append(['Value 1', 'Value 2', 'Value 3'])
    ```

- **셀 너비 설정**: `worksheet.column_dimensions['A'].width = 20`로 특정 열 너비 설정

    ```python
    worksheet.column_dimensions['A'].width = 20
    ```

- **폰트 설정**: `openpyxl.styles.Font`로 폰트 속성(굵기, 색상, 밑줄 등) 설정

    ```python
    from openpyxl.styles import Font

    font = Font(bold=True, color='FF0000', underline='single')
    cell.font = font
    ```

- **셀에 하이퍼링크 추가**: `cell.hyperlink = 'https://example.com'`로 특정 셀에 하이퍼링크 추가

    ```python
    cell.hyperlink = 'https://example.com'
    ```

- **셀 정렬 설정**: `openpyxl.styles.Alignment`로 셀 정렬(중앙, 왼쪽, 오른쪽, 위, 아래 등) 설정

    ```python
    from openpyxl.styles import Alignment

    alignment = Alignment(horizontal='center', vertical='top')
    cell.alignment = alignment
    ```

- **텍스트 줄 바꿈**: `Alignment(wrap_text=True)`로 셀 내의 텍스트 자동 줄 바꿈 설정

    ```python
    from openpyxl.styles import Alignment

    alignment = Alignment(wrap_text=True)
    cell.alignment = alignment
    ```

- **셀 테두리 설정**: `openpyxl.styles.Border`와 `openpyxl.styles.Side`로 셀 테두리 스타일 설정

    ```python
    from openpyxl.styles import Border, Side

    # 얇은 선 테두리 스타일
    thin_border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))
    cell.border = thin_border

    # 굵은 선 테두리 스타일
    thick_border = Border(left=Side(style='thick'), right=Side(style='thick'), top=Side(style='thick'), bottom=Side(style='thick'))
    cell.border = thick_border
    ```

- **셀 배경색 설정**: `openpyxl.styles.PatternFill`로 셀 배경색 설정

    ```python
    from openpyxl.styles import PatternFill

    fill = PatternFill(start_color='FF0000', end_color='FF0000', fill_type='solid')
    cell.fill = fill
    ```

````python
import requests , openpyxl
from bs4 import BeautifulSoup

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = 'board'

header = ['No','게시글제목','댓글']
excel_sheet.append(header)
excel_sheet.column_dimensions['A'].width = 20
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 80

res = requests.get('https://davelee-fun.github.io/trial/board/news')
soup = BeautifulSoup(res.content, 'html.parser')
titles = soup.select('div.list_item')
index =0

for title in titles[:5]:
    title_name = title.select_one('span.subject_fixed')
    title_replies = title.select_one('span.rSymph05')
    title_link = title.select_one('a.list_subject')
    default_url = 'https://davelee-fun.github.io/trial/board/'
    
    if title_name != None:
        index += 1
        title_data = [index ,title_name.get_text().strip(), '[' + title_replies.get_text() + ']개',"\n",default_url + title_link['href']]
        excel_sheet.append(title_data)
        
        #댓글 크롤링부분
        res_title = requests.get(default_url + title_link['href'])
        soup_title = BeautifulSoup(res_title.content, 'html.parser')
        reply_title = soup_title.select('div.comment_content')
        
        for reply in reply_title:
            reply_data = ["","",reply.get_text().replace("\n","").replace("\t","")]
            excel_sheet.append(reply_data)

excel_file.save("site.xlsx")
excel_file.close()
````

````python
# "requests" 라이브러리를 가져옵니다. 이 라이브러리를 통해 HTTP 요청을 보내고 웹 서버로부터의 응답을 처리할 수 있습니다.
import requests

# "bs4" 라이브러리에서 BeautifulSoup 함수를 가져옵니다. 이 라이브러리는 HTML과 XML 문서를 파싱하는데 사용됩니다. 이는 데이터를 쉽게 추출할 수 있도록 파싱 트리를 생성합니다.
from bs4 import BeautifulSoup

# requests.get 함수를 사용해 지정된 URL로 GET 요청을 보내고, 서버의 응답을 "res"라는 변수에 저장합니다.
res = requests.get('https://davelee-fun.github.io/trial/board/news.html')

# BeautifulSoup를 사용해 서버의 응답 내용을 파싱하고, 파싱된 데이터를 "soup"이라는 변수에 저장합니다.
# 'html.parser' 인자는 파싱을 파이썬의 내장 HTML 파서를 사용하여 수행하려는 것을 나타냅니다.
soup = BeautifulSoup(res.content, 'html.parser')

# select 메소드를 사용해 div 클래스 "list_item"을 가진 요소를 찾습니다. 이 메소드는 모든 일치하는 항목의 리스트를 반환합니다.
items = soup.select('div.list_item')

# "items" 리스트 내 각 일치하는 요소를 순회합니다.
for item in items:

    # select_one 메소드를 사용해 현재 "item" 내에 있는 span 클래스 "subject_fixed"을 가진 첫 번째 자식 요소를 찾습니다. 이 메소드는 첫 번째 일치 항목을 반환합니다.
    # 만약 요소가 발견되지 않으면 None이 반환됩니다.
    title = item.select_one('span.subject_fixed')

    # "title"이 발견되었는지 확인합니다.
    if title != None:

        # 만약 title이 발견되었다면, get_text 메소드를 사용해 HTML 요소 내의 텍스트를 추출하고, strip 메소드를 사용해 앞뒤 공백을 제거합니다.
        # 그리고 결과적인 title을 출력합니다.
        print (title.get_text().strip())
````
