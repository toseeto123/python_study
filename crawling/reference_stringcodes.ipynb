### 2단계: 회사명과 현재금액 한 줄에 출력해보기

````python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawling_stock_example')
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select('li.row_sty')


for inform in data:
    name = inform.select_one('div.st_name').get_text().replace("\n","").replace(" ","")
    price = inform.select_one('div.st_price').get_text().replace("\n","").replace(" ","")
    print(name,":",price ,"원")
````

````python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawling_stock_example.html')
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.select("li.row_sty")

for item in data:
    company = item.select_one('div.st_name').get_text().replace(" ", "").replace("\n", "")
    price = item.select_one('div.st_price').get_text().replace(" ", "").replace("\n", "")
    print (company, ", ", price)
````

### 3단계: 회사명, 현재금액, 변동률 한 줄에 출력해보기

````python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawling_stock_example')
soup = BeautifulSoup(res.content,'html.parser')
data = soup.select('li.row_sty')

for inform in data:
    name = inform.select_one('div.st_name').get_text().replace(" ","").replace("\n","")
    price = inform.select_one('div.st_price').get_text().replace(" ","").replace("\n","")
    rate = inform.select_one('div.st_rate').get_text().replace(" ","").replace("\n","")
    print("회사명:",name,":",price,"원",rate)
````

````python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawling_stock_example.html')
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.select("li.row_sty")

for item in data:
    company = item.select_one('div.st_name').get_text().replace(" ", "").replace("\n", "")
    price = item.select_one('div.st_price').get_text().replace(" ", "").replace("\n", "")
    rate = item.select_one('div.st_rate').get_text().replace(" ", "").replace("\n", "")
    print (company, ", ", price + "원, ", rate)
````
